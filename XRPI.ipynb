Getting molecular weight


```python
def mol_wt(seq):
  cleaned_seq = str(seq).replace("X", "")
  return ProteinAnalysis(cleaned_seq).molecular_weight()
rpi['mol_wt'] = rpi['protein_seq'].apply(mol_wt)
rpi.head()
```
